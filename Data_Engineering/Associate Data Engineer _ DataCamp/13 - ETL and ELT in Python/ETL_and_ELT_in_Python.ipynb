```python
raw_data = pd.read_csv(r'./Data/amazon_sales_cleaned_sql.csv')
raw_data.head(3)
```

```python
import logging

def transform(raw_data):
    raw_data["order_date"] = \
        pd.to_datetime(
            raw_data["order_date"],
            format="%m/%d/%y %H:%M"
        )

    clean_data = raw_data.loc[raw_data["price_each"] < 10, :]
    
    # Create an info log regarding transformation
    logging.basicConfig(
        filename='app.log',
        level=logging.info,
        format=f"Transformed 'Order Date' column to type 'datetime'."
    )
    
    # Create debug-level logs for the DataFrame before and after filtering
    logging.basicConfig(
        filename='app.log',
        level=logging.debug,
        format=f"Shape of the DataFrame before filtering: {raw_data.shape}"
    )
    # logging.debug(f"Shape of the DataFrame before filtering: {raw_data.shape}")
    # logging.debug(f"Shape of the DataFrame after filtering: {clean_data.shape}")

    logging.
    
    return clean_data
  
clean_sales_data = transform(raw_data)
```

```python
import logging

# Configure logging to write to a file
logging.basicConfig(
    filename='app.log',  # Log file name
    level=logging.DEBUG,  # Log level
    format='%(asctime)s - %(levelname)s - %(message)s'  # Log format
)

# Log messages
logging.debug("This is a debug message")
logging.info("This is an info message")
logging.warning("This is a warning message")
logging.error("This is an error message")
logging.critical("This is a critical message")
```
