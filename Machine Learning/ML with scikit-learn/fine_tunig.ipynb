# How Good is your Model

## Class Imbalance & Assessing Classification Performance

## Confusion Matrix

```python
import pandas as pd
import sklearn.metrics as skm
import sklearn.neighbors as skn
import sklearn.model_selection as skms

churn_df = pd.read_csv(r'..\Datasets\telecom_churn_clean.csv')
```

```python
churn_df.head()
```

```python
y = churn_df['churn'].values
X = churn_df[['account_length', 'customer_service_calls']].values
```

```python
knn = skn.KNeighborsClassifier(n_neighbors=7)
X_train, X_test, y_train, y_test = skms.train_test_split(X, y, test_size=0.4, random_state=42)
knn.fit(X_train, y_train)
y_pred = knn.predict(X_test)
```

```python
print(skm.confusion_matrix(y_test, y_pred))
```

```python
print(skm.classification_report(y_test, y_pred))
```
