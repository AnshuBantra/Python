```python
import pandas as pd

# Sample DataFrame
data = {
    'Year': [2024, 2024, 2024, 2024, 2024, 2024, 2025, 2025, 2025],
    'Month': [11, 11, 11, 12, 12, 12, 1, 1, 1],
    'Type': ['Weekdays', 'Saturday', 'Sunday', 'Weekdays', 'Sunday', 'Saturday', 'Weekdays', 'Saturday', 'Sunday'],
}

df = pd.DataFrame(data)

# Calculate value counts
counts = df.value_counts().reset_index(name='Count')

# Calculate totals for each Year-Month combination
totals = counts.groupby(['Year', 'Month'])['Count'].sum().reset_index()

# Add a 'Type' column with 'Total' label
totals['Type'] = 'Total'

# Concatenate the totals back to the original counts
df_with_totals = pd.concat([counts, totals], ignore_index=True)

# Sort the DataFrame by Year, Month, and a custom order for Type
df_with_totals['Type_order'] = df_with_totals['Type'].apply(lambda x: 1 if x == 'Total' else 0)
df_with_totals = df_with_totals.sort_values(by=['Year', 'Month', 'Type_order']).reset_index(drop=True)

# Drop the helper column used for sorting
df_with_totals = df_with_totals.drop(columns=['Type_order'])

print(df_with_totals)
```

```python
import pandas as pd

# Example data
data = {
    'Category1': ['A', 'A', 'B', 'B', 'C', 'C'],
    'Category2': ['X', 'Y', 'X', 'Y', 'X', 'Y'],
    'Category3': ['P', 'P', 'Q', 'Q', 'R', 'R'],
    'Value': [10, 20, 30, 40, 50, 60]
}

# Create a DataFrame
df = pd.DataFrame(data)

# Group by multiple columns and get the size of each group (i.e., count)
grouped = df.groupby(['Category1', 'Category2', 'Category3']).size().reset_index(name='Count')

# This will store the new DataFrame with totals inserted
result_df = pd.DataFrame(columns=grouped.columns)

# Iterate over each unique combination of Category1 and Category2
for (cat1, cat2), group in grouped.groupby(['Category1', 'Category2']):
    # Sort each group by 'Count' in descending order
    group_sorted = group.sort_values(by='Count', ascending=False)
    
    # Append the sorted group to the result DataFrame
    result_df = pd.concat([result_df, group_sorted], ignore_index=True)
    
    # Calculate the total count for the current (Category1, Category2) group
    total_count = group['Count'].sum()
    
    # Create a new row for the total
    total_row = pd.DataFrame({
        'Category1': [f'Total {cat1}'],
        'Category2': [cat2],
        'Category3': [''],
        'Count': [total_count]
    })
    
    # Append the total row to the result DataFrame
    result_df = pd.concat([result_df, total_row], ignore_index=True)

print(result_df)
```

```python
import pandas as pd

# Example data
data = {
    'Category': ['A', 'A', 'B', 'B', 'C', 'C'],
    'Value': [10, 20, 30, 40, 50, 60]
}

# Create a DataFrame
df = pd.DataFrame(data)

# This will store the new DataFrame with totals inserted
result_df = pd.DataFrame(columns=df.columns)

# Group by 'Category' and iterate over each group to calculate and insert totals
for category, gp in df.groupby('Category'):
    # Append the group to the result DataFrame
    result_df = pd.concat([result_df, gp], ignore_index=True)
    
    # Calculate the total for the current group
    total_value = group['Value'].sum()
    
    # Create a new row for the total
    total_row = pd.DataFrame({'Category': [f'Total {category}'], 'Value': [total_value]})
    
    # Append the total row to the result DataFrame
    result_df = pd.concat([result_df, total_row], ignore_index=True)

print(result_df)
```

```python
import pandas as pd

# Example data
data = {
    'Category': ['A', 'A', 'B', 'B', 'C', 'C'],
    'Value': [10, 20, 30, 40, 50, 60]
}

# Create a DataFrame
df = pd.DataFrame(data)

# This will store the new DataFrame with totals inserted
result_df = pd.DataFrame(columns=df.columns)

# Group by 'Category' and iterate over each group
for category, group in df.groupby('Category'):
    # Sort the group by 'Value' in descending order
    group_sorted = group.sort_values(by='Value', ascending=False)
    
    # Append the sorted group to the result DataFrame
    result_df = pd.concat([result_df, group_sorted], ignore_index=True)
    
    # Calculate the total for the current group
    total_value = group['Value'].sum()
    
    # Create a new row for the total
    total_row = pd.DataFrame({'Category': [f'Total {category}'], 'Value': [total_value]})
    
    # Append the total row to the result DataFrame
    result_df = pd.concat([result_df, total_row], ignore_index=True)

print(result_df)
```
