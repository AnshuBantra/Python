```python
# https://medium.com/learning-data/7-unique-pandas-functions-to-elevate-your-analysis-262fcb0db7be
import pandas as pd
```

```python
# Sample data
data1 = pd.DataFrame({
    'month': ['Jan', 'Feb', 'Marh', 'Aprh'],
    'value': [10, 20, 305, 405]
}, index = [1, 2, 3, 4])
data1
# extracted data
data2 = pd.DataFrame({
    'month': ['Aprh','Marh'],
    'value': [305, 405]
}, index=[3, 4])
data2
# Correcting the odd rows
data2 = pd.DataFrame({
    'month': ['Apr', 'Mar'],
    'value': [30, 40]
}, index=[3, 4])
data2
# Update the original data with corrected values
data1.update(data2)
data1
# APPLY
# Sample data
data = pd.DataFrame({
    'A': [1, 2, 3],
    'B': [4, 5, 6]
})
data.info()
data['A'] = data['A'].apply(lambda x: str(x) )
data.info()
# Creating Category

# Sample data
data = pd.DataFrame({
    'Age': [11, 16, 30, 51]
})

data['Category'] = data['Age'].apply( lambda x: \
                                        'Child' if x <= 12 else\
                                        'Teenager' if x <= 18 else\
                                        'Adult'
                                    )

display(data)
# Create new DataFrame from one

# Sample data
data = pd.DataFrame({
    'A': [1, 2, 3],
    'B': [4, 5, 6]
})

display(data)
# Apply a lambda function to each element
data_squared = data.apply(lambda x: x ** 2)

display(data_squared)
# Sample data
data = pd.DataFrame({
    'A': [10, 20, 30],
    'B': [1, 2, 3]
})
display(data)

def multiply(df):
    df['A'] *= 100
    return df

def add(df):
    df['B'] = df['B'] + 10
    return df

# Adding 10 to values in B and multiplying values in A with 100
data.pipe(multiply) \
    .pipe(add) \
    .pipe(display)
# CrossTab

data = pd.DataFrame({
    'Month': ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jan', 'May'],
    'Sales': [2000, 2200, 2100, 2500, 2700, 2000, 2000]
})
display(data)

# Set 'Month' as index
display(pd.crosstab(data['Month'], data['Sales']))

display(pd.crosstab(data['Month'], data['Sales']).reset_index())

display(pd.crosstab(data['Month'], data['Sales'], normalize='index'))

display(pd.crosstab(data['Month'], data['Sales'], normalize='columns'))
# pct_change --> Calculates the percent change between current and prior element

data = pd.DataFrame({
    'Month': ['Jan', 'Feb', 'Mar', 'Apr', 'May'],
    'Sales': [2000, 2200, 2100, 2500, 2700]
})

# Set 'Month' as index
data.set_index('Month', inplace=True)

# Calculate percentage change
data['Sales Change (%)'] = round((data['Sales'].pct_change() * 100), 2)

display(data)
# get_dummies

# Sample DataFrame with a categorical column
data = pd.DataFrame({
    'Answer': ['yes', 'no', 'no', 'yes', 'no']
})

display(data)
# Convert categorical column to dummy variables
dummies = pd.get_dummies(data['Answer'])

display(dummies)
# combine

# HR Data (data1)
data1 = pd.DataFrame({
    'EmployeeID': [101, 102, 103, 104],
    'Name': ['Alice', 'Bob', 'Charlie', 'David'],
    'Department': ['HR', 'Sales', 'HR', None],
    'Salary': [60000, 55000, 58000, 62000],
})
display(data1)

# Sales Data (data2)
data2 = pd.DataFrame({
    'EmployeeID': [101, 102, 104, 105],
    'SalesTarget': [50000, 70000, 65000, 80000],
    'SalesAchieved': [48000, 75000, 66000, 81000],
    'Department': ['HR', None, 'Sales', 'Sales']
})
display(data2)

# Combine using combine_first
combined_data = data1.set_index('EmployeeID')
combined_data.combine_first(data2.set_index('EmployeeID')).reset_index()

display(combined_data)

```
