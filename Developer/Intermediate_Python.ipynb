```python
# Basic scatter plot, log scale
plt.scatter(data=gapminder_df, x='gdp_cap', y='life_exp')
plt.xscale('log') 

# Strings
xlab = 'GDP per Capita [in USD]'
ylab = 'Life Expectancy [in years]'
title = 'World Development in 2007'

# Add axis labels
plt.xlabel(xlab)
plt.ylabel(ylab)

# Add title
plt.title(title)

# After customizing, display the plot
plt.show();
```

```python
# Scatter plot
plt.scatter(data=gapminder_df, x='gdp_cap', y='life_exp')

# Previous customizations
plt.xscale('log') 
plt.xlabel('GDP per Capita [in USD]')
plt.ylabel('Life Expectancy [in years]')
plt.title('World Development in 2007')

# Definition of tick_val and tick_lab
tick_val = [1000, 10000, 100000]
tick_lab = ['1k', '10k', '100k']

# Adapt the ticks on the x-axis
plt.xticks(tick_val, tick_lab)

# After customizing, display the plot
plt.show();
```

```python
# Import numpy as np
import numpy as np

# Store pop as a numpy array: np_pop
pop = list(gapminder_df['population'])
np_pop = np.array(pop)

# Double np_pop
np_pop = np_pop*2/1000000

# Update: set s argument to np_pop
plt.scatter(data=gapminder_df, x='gdp_cap', y='life_exp', s = np_pop, alpha=0.5)

# Previous customizations
plt.xscale('log') 
plt.xlabel('GDP per Capita [in USD]')
plt.ylabel('Life Expectancy [in years]')
plt.title('World Development in 2007')
plt.xticks([1000, 10000, 100000],['1k', '10k', '100k'])

# Display the plot
plt.show();
```

```python
colour_dict = {
  'Asia':'red',
  'Europe':'green',
  'Africa':'blue',
  'Americas':'yellow',
  'Oceania':'black'
}

gapminder_df['col'] = gapminder_df['cont']
gapminder_df.replace({'col':colour_dict}, inplace=True)
```

```python
gapminder_df.head()
```

```python
# Specify c and alpha inside plt.scatter()
plt.scatter(
    data = gapminder_df
    , x = 'gdp_cap'
    , y = 'life_exp'
    , s = np_pop
    , c = 'col'
    , alpha = 0.5)

# Previous customizations
plt.xscale('log') 
plt.xlabel('GDP per Capita [in USD]')
plt.ylabel('Life Expectancy [in years]')
plt.title('World Development in 2007')
plt.xticks([1000,10000,100000], ['1k','10k','100k'])

# Show the plot
plt.show();
```

```python
# Scatter plot
plt.scatter(data = gapminder_df
            , x = 'gdp_cap'
            , y = 'life_exp'
            , s = np_pop
            , c = 'col'
            , alpha = 0.6
)

# Previous customizations
plt.xscale('log') 
plt.xlabel('GDP per Capita [in USD]')
plt.ylabel('Life Expectancy [in years]')
plt.title('World Development in 2007')
plt.xticks([1000,10000,100000], ['1k','10k','100k'])

# Additional customizations
plt.text(1550, 71, 'India')
plt.text(5700, 80, 'China')

# Add grid() call
plt.grid(True)

# Show the plot
plt.show();
```
