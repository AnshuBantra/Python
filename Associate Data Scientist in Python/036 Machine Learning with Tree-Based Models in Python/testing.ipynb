```python
def expand_list(lst):
  expanded_list=[]
  for item in lst.split(', '):
      if '-' in item: 
        start, end = map(int, item.split('-'))
        expanded_list.extend(range(start, end + 1))
      else:
        expanded_list.append(int(item))
  return expanded_list
```

```python
def pig_latin_to_english(pig_latin_word):
    # Check if the word ends with "ay"
    if pig_latin_word.endswith("ay"):
        # Remove the "ay" suffix
        core_word = pig_latin_word[:-2]
        # Find the split point where the consonant cluster starts
        consonant_cluster_index = core_word.rfind(core_word[-1])
        # The consonant cluster comes after the first vowel
        # Split the word into the consonant cluster and the rest
        if consonant_cluster_index != -1:
            consonant_cluster = core_word[consonant_cluster_index:]
            rest_of_word = core_word[:consonant_cluster_index]
            return consonant_cluster + rest_of_word
    return pig_latin_word

# Examples
print(pig_latin_to_english("ankthay"))  # Output: thank
print(pig_latin_to_english("ouyay"))    # Output: you
print(pig_latin_to_english("ofay"))     # Output: of
```
