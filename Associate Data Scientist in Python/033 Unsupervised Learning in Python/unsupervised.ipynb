#### Wines

```python
df_wines = pd.read_csv(r'https://assets.datacamp.com/production/repositories/655/datasets/2b27d4c4bdd65801a3b5c09442be3cb0beb9eae0/wine.csv')
df_wines.head()
```

```python
df_wines.shape
```

```python
model = skcl.KMeans(n_clusters=3)
labels = model.fit_predict(df_wines.drop('class_name', axis=1))
labels
```

```python
df_test = pd.DataFrame({  'labels': labels,
                          'varieties': df_wines['class_name'].values
                      })
pd.crosstab(df_test['labels'], df_test['varieties'])
```

```python
import seaborn as sns
sns.relplot(  data=df_wines,
              x='od280',
              y='malic_acid',
              hue='class_name'
            )
```

```python
import sklearn.preprocessing as skpr
scaler = skpr.StandardScaler()
scaler.fit(df_wines.drop('class_name', axis=1))
skpr.StandardScaler(copy=True, with_mean=True, with_std=True)
samples_scaled = scaler.transform(df_wines.drop('class_name', axis=1))
samples_scaled
```

```python
import sklearn.pipeline as skpi
scaler = skpr.StandardScaler()
kmeans = skcl.KMeans(n_clusters=3)
pipeline = skpi.make_pipeline(scaler, kmeans)
pipeline.fit(df_wines.drop('class_name', axis=1))
labels = pipeline.predict(df_wines.drop('class_name', axis=1))
labels
```

```python
df_test = pd.DataFrame({  'labels': labels,
                          'varieties': df_wines['class_name'].values
                      })
pd.crosstab(df_test['labels'], df_test['varieties'])
```
